## Using StructType and StructField

```python
from pyspark.sql import *
from pyspark.sql.types import *
```

```python
data=[("Alice",28,5500.50),
      ("Bob",32,6000.00),
      ("Charlie",35,7000.00)
     ]
##StructType is nothing just a collection or list of StructField objects.

## StructField defines the metadata for a single column
schema=StructType([StructField("name",StringType(),True),  ##No Null//Null
                       StructField("age",IntegerType(),True),
                       StructField("salary",FloatType(),True)
                      ])
df=spark.createDataFrame(data,schema)
```

```python
df.printSchema()
```
