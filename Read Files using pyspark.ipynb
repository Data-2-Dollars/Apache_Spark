Reading CSV's

```python
df=spark.read.format("csv")\
    .option("header",True)\
    .option("inferSchema","true")\
        .load("/Volumes/bronze_layer/demo/pyspark_demo_vol/customers-100.csv")
df.display()
```

##Reading JSONs

```python
df_json=spark.read.format("JSON")\
    .option("inferSchema",True)\
    .option("mutliline",True)\
        .load('/Volumes/bronze_layer/demo/pyspark_demo_vol/Weather.json')

df_json.display()
```

##Reading Parquet

```python
df_parquet=spark.read.format("parquet")\
    .load("/Volumes/bronze_layer/demo/pyspark_demo_vol/weather.parquet")

df_parquet.display()
```

## Malformed Records

```python
df_json=spark.read.format("JSON")\
    .option("inferSchema",True)\
    .option("mutliline",True)\
        .option("mode","PERMISSIVE")\
        .load('/Volumes/bronze_layer/demo/pyspark_demo_vol/Weather.json')

df_json.display()
```

```python
df_json=spark.read.format("JSON")\
    .option("inferSchema",True)\
    .option("mutliline",True)\
        .option("mode","DROPMALFORMED")\
        .load('/Volumes/bronze_layer/demo/pyspark_demo_vol/Weather.json')

df_json.display()
```

```python
df_json=spark.read.format("JSON")\
    .option("inferSchema",True)\
    .option("mutliline",True)\
        .option("mode","FAILFAST")\
        .load('/Volumes/bronze_layer/demo/pyspark_demo_vol/Weather.json')

df_json.display()
```
