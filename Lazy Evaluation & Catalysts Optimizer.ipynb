```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
##LAZY EVALUATION

data=[
    (1,"Jaswinder","Data Engineer"),
     (2,"Sahil","Data Scientist"),
      (3,"Ayussh","Data Analyst"),
       (4,"Ravi","Data Engineer"),
           (5,"Akansha","Data enthusiast")
]
columns=["id","name","designation"]
df=spark.createDataFrame(data,columns)

df_new=df.filter(col("designation").like("%Engineer"))
```
