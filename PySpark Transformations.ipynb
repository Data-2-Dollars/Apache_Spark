##String Functions

```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

# Initialize Spark Session
spark = SparkSession.builder \
    .appName("PySparkStringFunctionsDemo") \
    .getOrCreate()

# Create sample data with messy spacing, mixed casing, and delimiters
data = [
    ("  daTa sCienCe  ", "Python,Spark,SQL", "ID-101-A"),
    ("  mAcHinE LeArNiNg  ", "R,Python,TensorFlow", "ID-202-B"),
    ("  aRtIfIcIaL InTeLlIgEnCe  ", "Java,C++,Python", "ID-303-C")
]

columns = ["raw_text", "skills", "code"]
df = spark.createDataFrame(data, columns)
```

```python
# --- 1. CASE CONVERSION & LENGTH ---
df_case = df.select(
    F.col("raw_text"),
    F.lower(F.col("raw_text")).alias("lower"),
    F.upper(F.col("raw_text")).alias("upper"),
    F.initcap(F.col("raw_text")).alias("initcap"),
    F.length(F.col("raw_text")).alias("length")
)
display(df_case)
```

```python

# --- 2. TRIMMING & WHITESPACE REMOVAL ---
# We use a helper column 'clean' to run subsequent functions effectively
df_clean = df.withColumn("clean", F.trim(F.col("raw_text")))

df_trim = df_clean.select(
    F.col("raw_text"),
    F.ltrim(F.col("raw_text")).alias("ltrim"),
    F.rtrim(F.col("raw_text")).alias("rtrim"),
    F.col("clean")
)

display(df_trim)
```

```python

# --- 3. CONCATENATION & SPLITTING ---
df_concat_split = df_clean.select(
    F.concat(F.col("clean"), F.lit(" Developer")).alias("concat"),
    F.concat_ws(" | ", F.col("clean"), F.col("code")).alias("concat_ws"),
    F.split(F.col("skills"), ",").alias("split_array")
)

display(df_concat_split)
```

```python

# --- 4. SUBSTRING & EXTRACTION ---
df_sub = df_clean.select(
    F.col("clean"),
    F.substring(F.col("clean"), 1, 4).alias("substring_1_4"),
    F.substring_index(F.col("code"), "-", 2).alias("sub_index_before"),
    F.substring_index(F.col("code"), "-", -1).alias("sub_index_after")
)
display(df_sub)

```

```python
# --- 5. SEARCHING & MATCHING ---
df_search = df_clean.select(
    F.col("clean"),
    F.col("clean").contains("Data").alias("contains_Data"),
    F.col("clean").startswith("daTa").alias("startswith_daTa"),
    F.col("clean").endswith("Ce").alias("endswith_Ce"),
    F.instr(F.col("clean"), "sCienCe").alias("instr_pos"),
    F.locate("LeArNiNg", F.col("clean")).alias("locate_pos")
)

display(df_search)
```

```python

# --- 6. PADDING & REPLACEMENT ---
df_pad_replace = df_clean.select(
    F.col("clean"),
    F.replace(F.col("clean"), F.lit("daTa"), F.lit("Big Data")).alias("replace"),
    F.regexp_replace(F.col("clean"), "[a-z]", "*").alias("regexp_replace"),
    F.lpad(F.col("clean"), 20, "-").alias("lpad"),
    F.rpad(F.col("clean"), 20, "*").alias("rpad")
)

df_pad_replace.display()
```

Window Functions

```python
data = [
    ("Alice", "Sales", 7000, "2026-01-01"),
    ("Bob", "Sales", 8000, "2026-01-02"),
    ("Charlie", "Sales", 8000, "2026-01-03"),
    ("David", "Sales", 9000, "2026-01-04"),
    ("Eve", "IT", 6000, "2026-01-01"),
    ("Frank", "IT", 6500, "2026-01-02"),
    ("Grace", "IT", 7000, "2026-01-03")
]

df = spark.createDataFrame(data, ["Name", "Department", "Salary", "Date"])

df.show()
```

##ROW_NUMBER, RANK AND DENSE_RANK

```python
window_spec=Window.partitionBy("Department").orderBy(desc("Salary"))

df.select("Name","Department","Salary",
          row_number().over(window_spec).alias("ROW_NUM"),
          rank().over(window_spec).alias("rnk"),
          dense_rank().over(window_spec).alias("DRNK"),
          ).show()

```

## LAG ^&^ LEAD

```python
jaswinder=Window.partitionBy("Department").orderBy("Date")

df.select("Name","Department","Salary","Date",
          lag("Salary",2).over(jaswinder).alias("LAG_COL"),
          lead("Salary",2).over(jaswinder).alias("Lead_COL"),
          ).show()
```

##Aggregation Window Functions

```python
jas_agg=Window.partitionBy("Department").orderBy("Date").rowsBetween(Window.unboundedPreceding,Window.currentRow)

df.select("Name","Department","Salary","Date",
        sum("Salary").over(jas_agg).alias("sum"),
          avg("Salary").over(jas_agg).alias("avg"),
          ).show()
```

##udf

mETHOD`1

```python
def cal_discount(price,discount_val):
    if price is None:
        return 0.0
    return price *(1-discount_val/100)
#registering the udf
discount_udf=udf(cal_discount,DoubleType())
```

```python
data = [("Laptop", 1200.0, 10), ("Mouse", 50.0, 5), ("Monitor", 300.0, 0)]
columns = ["Item", "Price", "Discount_Pct"]
df = spark.createDataFrame(data, columns)
```

```python
df_udf=df.withColumn("Final_Price",
              discount_udf(df["Price"],df["Discount_Pct"])       
                     )

df_udf.display()
```

Method 2 using udf with spark sql

```python
spark.udf.register("sql_udf",cal_discount,DoubleType())

df_sql_udf=df.withColumn("Final_Price",
              expr("sql_udf(Price,Discount_Pct)")       
                     )
df_sql_udf.display()
```
